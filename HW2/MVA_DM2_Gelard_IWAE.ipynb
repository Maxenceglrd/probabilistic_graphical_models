# Specifying our graphical model

We will use a **deep latent variable model with a Gaussian prior and a Bernoulli observation model**. This can be written:

$$p_{\boldsymbol{\theta}}(\mathbf{x}_1,...,\mathbf{x}_n) = \prod_{i=1}^n p(\mathbf{x}_i|\mathbf{z}_i)p(\mathbf{z}_i),$$
$$p(\mathbf{z}_i) = \mathcal{N}(\mathbf{z}_i|\mathbf{0}_d,\mathbf{I}_d), $$
$$p_{\boldsymbol{\theta}}(\mathbf{x}_i|\mathbf{z}_i) = \mathcal{B} (\mathbf{x}_i|\boldsymbol{\pi}_{\boldsymbol{\theta}}(\mathbf{z}_i)),$$

where $\boldsymbol{\pi}_{\boldsymbol{\theta}}: \mathbb{R}^d \rightarrow \mathbb{[0,1]}^p$ is a function (called the **decoder**) parametrised by a deep neural net. This neural net will be defined as a keras sequential model.

```python
d = 10 # dimension of the latent space

p_z = tfd.Independent(
    tfd.Normal(loc=tf.zeros(d, tf.float32), scale=tf.ones(d, tf.float32)),
    reinterpreted_batch_ndims=1
)  # that's the prior
```

```python
h = 100 # number of hidden units

sigma = "relu"

decoder = tfk.Sequential([
  tfkl.InputLayer(input_shape=[None, d]),
  tfkl.Dense(h, activation=sigma),
  tfkl.Dense(h, activation=sigma),
  tfkl.Dense(28*28)
])

# changed input dim from [d, ] to [None, d] as we will
# have tensors of shape [K, batch_size, d], so it will
# avoid too many reshape operations 
```

Note that the decoder needs to output probabilities, which can be constrained using a sigmoid final layer. However, it is generally better (i.e. computationnaly more stable) to work with [logits](https://stackoverflow.com/questions/34240703/what-are-logits-what-is-the-difference-between-softmax-and-softmax-cross-entrop) than probabilities, when doing deep learning. For instance, all the distributions of TensorFlow probability or Pytorch distributions that use probabilities as parameters (e.g. [Bernoulli](https://www.tensorflow.org/probability/api_docs/python/tfp/distributions/Bernoulli), [categorical](https://www.tensorflow.org/probability/api_docs/python/tfp/distributions/Categorical) or [binomial](https://www.tensorflow.org/probability/api_docs/python/tfp/distributions/Binomial)) can use $\texttt{logits}$ or $\texttt{probs}$ as arguments. It is generally better to use  $\texttt{logits}$. This is why this keras model for the decoder does not output probabilities, but logits.





**1. Implement a function that performs ancestral sampling for this deep latent variable model, and show 5 sampled images from the initialised model. Of course, we expect these samples to look like random noise, since we have not trained our model yet.**

```python
def ancestral_sampling(decoder, prior, n_samples):
    prior_samples = prior.sample(sample_shape=n_samples)
    model_logits = decoder(prior_samples)
    posterior = tfd.Bernoulli(logits=model_logits)
    model_samples = posterior.sample()
    return model_samples
```

```python
def plot_sampled_images():
    deep_model_samples = np.array(ancestral_sampling(decoder, p_z, 5))

    fig, ax = plt.subplots(1, 5, figsize=(18, 16))
    for i in range(5):
        plt.subplot(1, 5, i+1)
        plt.imshow(np.array(deep_model_samples[i, :]).reshape(28, 28), cmap='gray')
        plt.title(f"Sample nÂ°{i+1}")
        
plot_sampled_images()
```

## Encoder

To be able to train our model, we will need an **encoder** (aka **inference network**), that will allow us to approximate the intractable posterior $p(\mathbf{z}|\mathbf{x})$.
The approximate posterior is defined as follows
$$ q(\mathbf{z}|\mathbf{x}) = \mathcal{N}(\mathbf{z} | \mathbf{m}_\boldsymbol{\gamma} ( \mathbf{x} ), \text{Diag}(\mathbf{s}_\boldsymbol{\gamma} ( \mathbf{x} )),
$$ 
where $\mathbf{x} \mapsto (\mathbf{m}_\boldsymbol{\gamma} ( \mathbf{x} ), \mathbf{s}_\boldsymbol{\gamma} ( \mathbf{x} ))$ is a function from the data space (i.e. $\{0,1\}^p$) to $\mathbb{R}^d \times [0, \infty[^d$ parametrised by a deep neural network. In other words, the encoder outputs the mean and the diagonal of the covariance matrix of the approximate posterior $q(\mathbf{z}|\mathbf{x})$.

```python
encoder = tfk.Sequential([
  tfkl.InputLayer(input_shape=[28*28,]),
  tfkl.Dense(h, activation=sigma),
  tfkl.Dense(h, activation=sigma),
  tfkl.Dense(2*d) # 2*d because we need both the mean and the diagonal of the covariance of q(z|x)
])
```

## The IWAE objective

The **IWAE objective** is defined as
$$
\mathcal{L}_K (\boldsymbol{\theta,\gamma}) = \sum_{i=1}^n \mathbb{E}_{\mathbf{z}_{i1},\ldots,\mathbf{z}_{iK} \sim q_{\boldsymbol{\gamma}}(\mathbf{z}|\mathbf{x}_i)} \left[ \log\frac{1}{K} \sum_{k=1}^K \frac{p_{\boldsymbol{\theta}}(\mathbf{x}_i|\mathbf{z}_{ik}){p(\mathbf{z}_{ik})}}{{q_{\boldsymbol{\gamma}}(\mathbf{z}_{ik}|\mathbf{x}_i)}} \right]
$$

**2. Create a function that computes an unbiased estimate of the IWAE bound, and optimise it to train the model. You are allowed to use TF probability to avoid implementing the reparametrisation trick, and you are allowed to use keras for training.** 

Below is given the `IWAEModel` class which implements the defined deep latent variable model. It namely provides a function `estimate_IWAE_bound` that computes, as asked, un unbiased estime of the IWAE bound. This is done by Monte-Carlo sampling, i.e for each datapoint $x$, we sample $K$ ($K$ being a hyperparameter for the `IWAEModel`) $z_k$, in order to average the $K$ values of $\frac{p_{\boldsymbol{\theta}}(\mathbf{x}_i|\mathbf{z}_{ik}){p(\mathbf{z}_{ik})}}{{q_{\boldsymbol{\gamma}}(\mathbf{z}_{ik}|\mathbf{x}_i)}}$.

In order to train the model, we define the loss being the opposite of the so-defined unbiased estimate of the IWAE bound. This loss is miinmized using Tensorflow autodifferentation and Adam optimizer.

```python
class IWAEModel(tfk.Model):
    def __init__(self, encoder, decoder, prior, K, latent_dim):
        super(IWAEModel, self).__init__()
        
        self.encoder = encoder
        self.decoder = decoder
        self.prior = prior
        self.K = K
        self.latent_dim = latent_dim
        
        self.optimizer = tfk.optimizers.Adam(1e-3)
        

    def estimate_IWAE_bound(self, X):
        n = X.shape[0]
        mean, covariance = tf.split(
            self.encoder(X), num_or_size_splits=2, axis=1
        )
        valid_covariance = tf.exp(covariance) # variances (diagonal elements) must not be negative
        # this trick is used in the provided TF tutorial on convolutional VAEs.
        q_z_x = tfd.Independent(
            tfd.Normal(loc=mean, scale=valid_covariance, validate_args=True),
            reinterpreted_batch_ndims=-1
        )

        # q_gamma(z_ik | x_i)
        z_given_x = q_z_x.sample(sample_shape=self.K)
        log_q_z_given_x = q_z_x.log_prob(z_given_x)
        log_q_z_given_x = tf.reduce_sum(log_q_z_given_x, axis=-1)
        # we are summing over the batch dimension

        # p(z_ik)
        log_prior = self.prior.log_prob(z_given_x)
        
        # p_theta(x_i | z_ik)
        model_logits = self.decoder(z_given_x)
        posterior = tfd.Bernoulli(logits=model_logits)
        log_p_x_given_z = posterior.log_prob(X)
        log_p_x_given_z = tf.reduce_sum(log_p_x_given_z, axis=-1)

        # all "log_p" shape is [self.K, n] 

        # Compute loss
        sum_log_prob_over_K = tf.reduce_logsumexp(
            log_p_x_given_z + log_prior - log_q_z_given_x, axis=0
        )
        return tf.reduce_mean(sum_log_prob_over_K)

        
    @tf.function
    def train(self, X_train):
        with tf.GradientTape(persistent=True) as tape:
            iwae_loss = - self.estimate_IWAE_bound(X_train)
        gradients = tape.gradient(iwae_loss, self.trainable_variables)
        self.optimizer.apply_gradients(zip(gradients, self.trainable_variables))
        return float(iwae_loss)
```

```python
def train_model(model, train_dataset, batch_size, n_epochs):
    for i in range(n_epochs):
        iwae_loss = 0
        for X_train in train_dataset:
            iwae_loss += model.train(X_train)
        if i % 50 == 0:
            print(f"Epoch: {i + 1} / {n_epochs}, Loss: {iwae_loss / batch_size}")
```

```python
K = 20
model = IWAEModel(encoder, decoder, p_z, K, d)
```

```python
# This dataloading is the one used in the link which was given
# at the beggining of the assignment: https://www.tensorflow.org/tutorials/generative/cvae

train_size = len(train_images)
test_size = len(test_images)
batch_size = 2048

train_dataset = (tf.data.Dataset.from_tensor_slices(train_images)
                 .shuffle(train_size).batch(batch_size))
test_dataset = (tf.data.Dataset.from_tensor_slices(test_images)
                 .shuffle(test_size).batch(batch_size))
```

```python
training = False
if training:
    train_model(model, train_dataset, batch_size, n_epochs=1000)
else:
    if K == 20:
        model.load_weights("weights/my_model_6")
    elif K == 5:
        model.load_weights("weights/my_model_7")
    else:
        print("No model with this value of K has been trained, please change training boolean to True")
```

Below are computed the values of the train/test IWAE bounds after training

```python
train_IWAE_bound = 0
test_IWAE_bound = 0
for X_train in train_dataset:
    train_IWAE_bound += model.estimate_IWAE_bound(X_train)
for X_test in test_dataset:
    test_IWAE_bound += model.estimate_IWAE_bound(X_test)

train_IWAE_bound /= train_size
test_IWAE_bound /= test_size

print(f"Train IWAE bound: {train_IWAE_bound}")
print(f"Test IWAE bound: {test_IWAE_bound}")
```

**4. Importance sampling and visualisation**

In this section, we are going to estimate the posterior mean of each datapoint, which will allow us to visualize the embedding of the datapoint in the latent space. As the latent space's dimension is $d=10$, we are going to visualize:
- Random pairs of dimensions of the embedding.
- The results of a PCA with 2 components of the embedding.

Two estimates are tried: first a simple Monte Carlo estimate, and then an importance sampling estimate that uses a StudentT distribution as proposal.

For the importande sampling estimate, the StudentT distribution for the proposal will have the same mean and variance as for $q(\mathbf{z}|\mathbf{x})$ ; and the degree of freedom has been set to 3. The importance sampling is derived by:

$\mathbb{E}(z|x_i) = \displaystyle\int z p(z|x_i) \, \mathrm{d}z \approx \displaystyle\int z q(z|x_i) \, \mathrm{d}z= \displaystyle\int \frac{z q(z|x_i)}{p_{student}(z|x_i)} p_{student}(z|x_i) \, \mathrm{d}z \approx \displaystyle \frac{1}{n_{samples}}\sum_{k=1}^{n_{samples}} \frac{z_k q(z_k|x_i)}{p_{student}(z_k|x_i)}$, for $z_1, ... z_{n_samples} \sim p_{student}$.

```python
def importance_sampling_estimate(x, n_samples=500):
    n = x.shape[0]
    x = tf.reshape(x, [n, 28 * 28])
    mean, covariance = tf.split(encoder(x), num_or_size_splits=2, axis=1)
    valid_covariance = tf.math.softplus(covariance)
    q_z_x = tfd.Normal(loc=mean, scale=valid_covariance, validate_args=True)
    proposal = tfp.distributions.StudentT(
      df=3, loc=mean, scale=valid_covariance, validate_args=True
    )
    
    z_given_x = proposal.sample(sample_shape=n_samples)
    prob_q_z_x = q_z_x.prob(z_given_x)
    prob_q_z_x = tf.reduce_mean(prob_q_z_x, axis=-1)

    proposal_prob = proposal.prob(z_given_x)
    proposal_prob = tf.reduce_mean(proposal_prob, axis=-1)

    prob_coef = prob_q_z_x / proposal_prob
    prob_coef_repeat = tf.reshape(tf.repeat(prob_coef, d, axis=-1), [n_samples, n, d])
    estimation = tf.reduce_mean(z_given_x * prob_coef_repeat, axis=0)
    variance = tf.reduce_mean(tf.square(z_given_x - estimation), axis=0)
    return estimation, variance


def monte_carlo_estimate(x, n_samples=500):
    n = x.shape[0]
    x = tf.reshape(x, [n, 28 * 28])
    mean, covariance = tf.split(encoder(x), num_or_size_splits=2, axis=1)
    valid_covariance = tf.math.softplus(covariance)
    q_z_x = tfd.Normal(loc=mean, scale=valid_covariance, validate_args=True)
    z_given_x = q_z_x.sample(sample_shape=n_samples)
    estimation = tf.reduce_mean(z_given_x, axis=0)
    variance = tf.reduce_mean(tf.square(z_given_x - estimation), axis=0)
    return estimation, variance
```

```python
dimension_pairs = [(0, 1), (1, 2), (3, 4), (7, 9)]
posterior_monte_carlo, variance_monte_carlo = monte_carlo_estimate(train_images, n_samples=100)
posterior_importance_sampling, variance_is = importance_sampling_estimate(train_images, n_samples=100)

fig, ax = plt.subplots(4, 2, figsize=(16, 18))
for (i, (k, l)) in enumerate(dimension_pairs):
    plt.subplot(4, 2, 2*i+1)
    plt.scatter(
        posterior_monte_carlo[:, k],
        posterior_monte_carlo[:, l],
        c=y_train,
    )
    plt.title(f"Posterior mean - MC, axis {k}, {l}")

    plt.subplot(4, 2, 2*i+2)
    plt.scatter(
        posterior_importance_sampling[:, k],
        posterior_importance_sampling[:, l],
        c=y_train,
    )
    plt.title(f"Posterior mean - Importance Sampling, axis {k}, {l}")
```

```python
mean_mc_variance = tf.reduce_mean(variance_monte_carlo)
mean_is_variance = tf.reduce_mean(variance_is)

print(f"Variance of MC estimator: {float(mean_mc_variance)}")
print(f"Variance of Important Sampling estimator: {float(mean_is_variance)}")
```

Even if taken separately, each class (i.e each digit) does not have a posterior distribution which looks like a normal distribution, they jointly have a distribution which is not so far from a normal distribution. Unexpectedly, the use of the importance sampling didn't succeed in reducing the variance of the estimation that we got using a simple Monte Carlo estimation.

```python
pca = PCA(n_components=2)
reduced_mc = pca.fit_transform(posterior_monte_carlo)

pca_is = PCA(n_components=2)
reduced_is = pca_is.fit_transform(posterior_importance_sampling)

fig, ax = plt.subplots(1, 2, figsize=(16, 8))

plt.subplot(1, 2, 1)
plt.scatter(reduced_mc[:, 0], reduced_mc[:, 1], c=y_train)
plt.title("Posterior mean with Monte Carlo estimate - PCA")

plt.subplot(1, 2, 2)
plt.scatter(reduced_is[:, 0], reduced_is[:, 1], c=y_train)
plt.title("Posterior mean with Importance Sampling estimate - PCA")
```
